```python

import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import confusion_matrix
from imblearn.metrics import classification_report_imbalanced
```

```python

# Load the data
file_path = Path('Resources/2019loans.csv')
df = pd.read_csv(file_path)

df.head()
```

# Split into Training and Testing

```python
# Create our features
X= df.drop(columns="loan_status")

# Create our target
y=df["loan_status"]
X
```

```python
#print(df.columns)
dummy = pd.get_dummies(df)
dummy.head()
```

```python
# Create our features
X= dummy.drop(columns="loan_status_low_risk")

# Create our target
y=dummy["loan_status_low_risk"]
X
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, 
                                                   y, 
                                                   random_state=1, 
                                                   stratify=y)
X_train.shape
```

```python
scaler = StandardScaler()

# Fitting the scaler
X_scaler = scaler.fit(X_train)

# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# RandomForestClassifier model

```python



# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier

brf = BalancedRandomForestClassifier(n_estimators=1000, random_state=1)
brf.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
y_pred_brf = brf.predict(X_test_scaled)
bas_brf=balanced_accuracy_score(y_test, y_pred_brf)
print(bas_brf)
```

```python
# Display the confusion matrix
cm_brf = confusion_matrix(y_test, y_pred_brf)
cm_df_brf = pd.DataFrame(
    cm_brf, index=["Actual High Risk", "Actual Low Risk"], columns=["Predicted High Risk", "Predicted Low Risk"]
)
cm_df_brf
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred_brf))
```
